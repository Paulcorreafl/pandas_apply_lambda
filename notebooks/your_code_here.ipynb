```python
import pandas as pd
```

```python
import this
```

```python
 Movie_Data = pd.read_csv("../data/input/IMDB-Movie-Data.csv")
Movie_Data
```

```python
#challenge1
def create_bin(x):
    if x in range(0,1000):
        return "cat 1"
    elif x in range(999,10000):
        return  "cat 2"
    elif x in range(9999,100000):
        return "cat 3"
    elif  x in range(99999,1000000):
        return "cat 4"
    elif x >= 1000000:
        return "cat 5"
```

```python
Movie_Data["bin"]= Movie_Data.apply(lambda row:create_bin(row["Votes"]), axis = 1)
Movie_Data
```

```python
#challenge2
Movie_Data["Revenue per minute"]= Movie_Data.apply(lambda row:(round(row["Revenue (Millions)"]/row["Runtime (Minutes)"],2)), axis = 1)
Movie_Data
```

```python
#challenge3
def rates(x, y):
    if "Thriller" in x:
        return y +1
    elif "Comedy" in x:
        return y - 1
    else:
        return y
   
```

```python
Movie_Data["New rating"]= Movie_Data.apply(lambda row:rates(row["Genre"],row["Rating"]), axis = 1)
Movie_Data.head()
```

```python
#challenge4
def check(x):
    if x < 2:
        return False
    elif x == 2:
        return True
    elif x > 2:
        for i in range(2, x):
            if x % i == 0:
                return False
        return True
```

```python
Movie_Data["Weird Prime check"]= Movie_Data.apply(lambda row:(check(int((sum(ord(c) for c in row["Actors"]))/row["Votes"]))), axis = 1)
Movie_Data.head()
```
